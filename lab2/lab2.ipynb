```python
from fuzzy import FuzzyVariable
from fuzzy import FuzzyCalculator
```

<h1>Задание</h1>
<p>
    Задача кондиционера – поддерживать оптимальную температуру
воздуха в комнате (t), охлаждая его, когда жарко, и нагревая, когда холодно.
Пусть, изменяя скорость вращения вентилятора (V), прогоняющего воздух
через охлаждающий элемент, мы можем менять температуру воздуха, тогда
алгоритм работы кондиционера может быть задан следующими правилами:
</p>
<table>
    <thead>
        <th style="text-align:center">№ правила</th>
        <th style="text-align:center">антецедент</th>
        <th style="text-align:center">консеквент</th>
    </thead>    
    <tbody>
        <tr>
            <td style="text-align:center">1</td>
            <td style="text-align:left">t = "высокая"</td>
            <td style="text-align:left">V = "высокая"</td>
        </tr>
        <tr>
            <td style="text-align:center">2</td>
            <td style="text-align:left">t = "средняя"</td>
            <td style="text-align:left">V = "средняя"</td>
        </tr>
        <tr>
            <td style="text-align:center">3</td>
            <td style="text-align:left">t = "низкая"</td>
            <td style="text-align:left">V = "низкая"</td>
        </tr>
    </tbody>    
</table>
<p>
    Разработать ПО, выполняющее расчеты скорости вращения
вентилятора V в зависимости от температуры воздуха t для пяти итераций.
</p>
<p>
    Значение начальной температуры (температуры в первой итерации) для
индивидуального варианта:
</p>
<p>
  t<sub>0</sub> = 12 + 0.5iС˚, где i - номер варианта.
</p>
<p>
  Для оставшихся итераций температуру принять по формуле:
</p>
<p>
  t<sub>текущей итерации</sub> = t<sub>предыдущей итерации</sub> + 0.3С˚.
</p>

```python
results = []
initialTemperature = 12 + 2 * 0.5 #2й вариант
temperatures = []
for i in range(0, 5):
    initialTemperature += 0.3
    temperatures.append(round(initialTemperature, 2))
```

```python
for temperature in temperatures:
    temperatures_set = FuzzyVariable([0, 20], [10, 30], [20, 60])
    speeds = FuzzyVariable([0, 400], [200, 600], [400, 1000])
    calculator = FuzzyCalculator(temperatures_set, speeds)
    output = calculator.calculate(temperature)
    print('Необходимая скорость вращения:', output, 'об/мин, при температуре', temperature, '°C')
    results.append(output)
```

```python
calculator.plot_output_graph(temperatures, results)
```
